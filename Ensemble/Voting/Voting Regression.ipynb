```python
from sklearn.datasets import load_diabetes
X,y=load_diabetes(return_X_y=True)
print(X.shape,y.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
svr=SVR()
```

```python
estimators=[('lr',lr),('dt',dt),('svr',svr)]
```

```python
import numpy as np
```

```python
for estimator in estimators:
    x=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(estimator[0],np.round(np.mean(x),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr=VotingRegressor(estimators)
x=cross_val_score(vr,X,y,cv=10,scoring='r2')
print("Voting Regressor",np.round(np.mean(x),2))
```

```python
l=[]
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            vr=VotingRegressor(estimators,weights=[i,j,k])
            x=cross_val_score(vr,X,y,cv=10,scoring='r2')
            l.append(np.round(np.mean(x),2))
            print(f"for 'i'={i}, 'j'={j}, 'k'={k}",np.round(np.mean(x),2))
print(max(l))
```

<h3>Voting Regressor Example.
VR can be used with same model with different paramters and it might give better answer than all models.</h3>

```python
dr1=DecisionTreeRegressor(max_depth=1)
dr1=DecisionTreeRegressor(max_depth=2)
dr3=DecisionTreeRegressor(max_depth=3)
dr4=DecisionTreeRegressor(max_depth=4)
dr5=DecisionTreeRegressor(max_depth=None)
```

```python
estimators=[('dr1',dr1),('dr2',dr2),('dr3',dr3),('dr4',dr4),('dr5',dr5)]
for estimator in estimators:
    x=cross_val_score(estimator[1],X,y,cv=10,scoring='r2')
    print(estimator[0],np.round(np.mean(x),2))
```

```python
vr=VotingRegressor(estimators)
x=cross_val_score(vr,X,y,cv=10,scoring='r2')
print("Voting Regressor",np.round(np.mean(x),2))
```
